```python
#Schumacher's Golden Era: Please filter the data to show me only the races won by Ferrari in 2004. We only want to see the schedule for that year.#

year_grp=df.groupby(['year'])
best_year=year_grp.get_group(2004)

team_grp=best_year.groupby(['team'])
ferrari=team_grp.get_group('Ferrari')
ferrari

#This Table Shows Schumacher's Golden Era.#
```

```python
#What is the total number of wins McLaren has achieved throughout its entire history?#
filt = df[df['team'].str.contains('McLaren')]
len(filt)
```

```python
#How do you know that McLaren doesn't just have one name, but many?#
all_teams= df['team'].unique()
mclaren = [team for team in all_teams if "McLaren" in str(team)]
mclaren
```

```python
#The 5 Kings in Monaco Programmatically#
monaco_circiut = df.groupby(['circuit'])
monaco= monaco_circiut.get_group('Circuit de Monaco')
bar= monaco['winner_name'].value_counts()
monaco_states = bar.head()
monaco_states = monaco_states.sort_values(ascending = True)
```

```python
#The 5 kings in Monaco Visually#
plt.barh(monaco_states.index , monaco_states.values , color = 'firebrick')
plt.title('The 5 kings of "Circuit de Monaco"')
plt.xlabel('Numeber of win')
plt.ylabel('Winners_Names')
plt.grid(True)
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
team_grp=df.groupby(['team'])
ferrari=team_grp.get_group('Ferrari')
ferrari_grp_=ferrari.groupby(['year'])
ferrari_standing= ferrari_grp_.count()
```

```python
plt.plot(ferrari_standing.index , ferrari_standing.values , color ='red' , marker = 'o' , linewidth ='1' )
plt.title('The Wins Of Ferrari of Hestori in F1')
plt.xlabel('Years')
plt.ylabel('Number of Wins')
plt.grid(True , alpha = 0.3)
plt.figure(figsize=(14, 6))
plt.style.use('fivethirtyeight')
```
